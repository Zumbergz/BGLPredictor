## 3. $L_2$-regularised least squares solution
### 3.1 Implementation
Now, you will need to implement the function `l2_rls_train` below. It should train a linear model by minimising the $L_2$-regularised sum of squares loss through zeroing the loss gradient. It should support both single-output and multi-output cases. This function should take a set of training samples and a user-specified regularisation parameter $\lambda$ as the input, and return the predicted weights. When $\lambda=0$, use pseudo inverse to implement the solution.

```python
def l2_rls_train(data, labels, hyperparam=1):
    """
    Calculate prediction weights using L2 regularised least squares.
    First pad data X with a column of ones, then calculate optimal weight vector w.
    In the case where hyperparam=0, use pseudoinverse instead to calculate the value of w.

    data: np matrix training data to train the model
    labels: corresponding vector labels to the data
    hyperparam: lamba value to be used in regularisation.

    Returns: a vector of weight values used for prediction
    """
    
    # This is just to be consistent with the lecture notes.
    X, y = data, labels
    N = X.shape[0]
    N2 = X.shape[1]
    # Expand X with a column of ones.
    X_tilde = np.c_[np.ones(N), X]
    X_tilde_mult =  np.transpose(X_tilde) @ X_tilde
    ident = np.eye(N2+1) * hyperparam
    if hyperparam == 0:
        w = np.linalg.pinv(X_tilde)@y
    else:
        inverse = np.linalg.inv(X_tilde_mult + ident)
        outer = np.transpose(X_tilde) @ y
        w = inverse @ outer
    
    # Return model parameters.
    return w
```

#### Experiment 1: Binary Classification

Start with a binary classification task to classify face images for subjects “1” and “30”.

Train three classifiers:
1. Set $y=-1$ for samples from “1” and $y=+1$ for samples from “30” for training a linear model, and use the threshold $T= 0$ to obtain the class prediction.
2. Set $y=0$ for samples from “1” and $y=1$ for samples from “30” for training a linear model, and use the threshold $T= 0.5$ to obtain the class prediction.
2. Set $y=1$ for samples from “1” and $y=30$ for samples from “30” for training a linear model, and use the threshold $T= 15.5$ to obtain the class prediction.

```python
def train_and_test_classifier(labels_train, labels_test, T, lab1, lab2):
    w1 = l2_rls_train(tr_data, labels_train, 0)
    #Calculate weights
    train_predictions = np.ones(len(tr_data))
    test_predictions = np.ones(len(te_data))
    #Initialise vectors to hold prediction values
    for i in range(len(tr_data)):
        predicts_train = l2_rls_predict(w1, tr_data[i,:])
        #If less than treshold, first label. Else, second label predicted
        if predicts_train < T:
            train_predictions[i] = lab1
        else:
            train_predictions[i] = lab2
    for i in range(len(te_data)):
        predicts_test = l2_rls_predict(w1, te_data[i,:])
        if predicts_test < T:
            test_predictions[i] = lab1
        else:
            test_predictions[i] = lab2
    #Calculate number predicted correctly for train and test sets
    correct_test = np.sum(test_predictions == labels_test)
    correct_train = np.sum(train_predictions == labels_train)
    test_error = 1 - correct_test / labels_test.size
    train_error = 1 - correct_train / labels_train.size
    return train_error, test_error

#Initialise vectors to hold training and testing errors for each classifier
train_errors_1 = np.ones(50)
test_errors_1 = np.ones(50)
train_errors_2 = np.ones(50)
test_errors_2 = np.ones(50)
train_errors_3 = np.ones(50)
test_errors_3 = np.ones(50)
for i in range(50):
    tr_ind, te_ind = partition_data(labels, num_per_class=3)
    #Keep all indices where the label is 1 or 30 as we are predicting for these classes
    tr_ind_1 = tr_ind[(labels[tr_ind]==1)| (labels[tr_ind]==30)]
    te_ind_1 = te_ind[(labels[te_ind]==1)| (labels[te_ind]==30)]
    tr_labels = labels[tr_ind_1]
    te_labels = labels[te_ind_1]
    tr_data = data[tr_ind_1, :]
    te_data = data[te_ind_1, :]
    #Set up different training labels for the three different classifiers.
    tr_labels_1 = np.where(tr_labels==1,-1, tr_labels)
    tr_labels_1 = np.where(tr_labels==30, 1, tr_labels_1)
    tr_labels_2 = np.where(tr_labels==1,0, tr_labels)
    tr_labels_2 = np.where(tr_labels==30, 1, tr_labels_2)
    tr_labels_3 = tr_labels
    #Set up different testing labels for 3 classifiers
    te_labels_1 = np.where(te_labels==1,-1, te_labels)
    te_labels_1 = np.where(te_labels==30, 1, te_labels_1)
    te_labels_2 = np.where(te_labels==1,0, te_labels)
    te_labels_2 = np.where(te_labels==30, 1, te_labels_2)
    te_labels_3 = te_labels
    #Place returned training and testing errors into earlier initialised vectors
    train_errors_1[i], test_errors_1[i] = train_and_test_classifier(tr_labels_1, te_labels_1, 0, -1, 1)
    train_errors_2[i], test_errors_2[i] = train_and_test_classifier(tr_labels_2, te_labels_2, 0.5, 0, 1)
    train_errors_3[i], test_errors_3[i] = train_and_test_classifier(tr_labels_3, te_labels_3, 15.5, 1, 30)
```

#### Experiment 2: Multi-class Classification

Now, proceed to the multi-class classification task to classify face images for all the 40 subjects.

Split your data to two sets: the training set contains 5 randomly selected images per subject and the test set contains the remaining images. The training set is used to train the model and select hyper-parameter $\lambda$, while the test set is only for final assessment of the trained classifier with the selected $\lambda$. 

Design an appropriate and complete machine learning experiment, which should include the training, hyper-parameter selection and evaluation stages. You can choose from the random subsampling, $k$-fold CV and LOO approaches for hyper-parameter selection. In the evaluation, you should construct a $40\times 40$ classification error matrix (confusion matrix) for test samples. 

Write your code below and make sure it compiles. 

```python
# Your code goes here
# Note that you may find it helpful to use one-hot encoding to deal with multi-class labels
tr_ind_mult, te_ind_mult = partition_data(labels, num_per_class=5)

# Set up one hot encoding labels for later use
maxi = np.max(labels)
ohe_labels = np.eye(maxi)[labels-1]
```

```python
#Select 3/4 of the labels used for training, rest for testing in each iteration
num_rand = (len(tr_ind_mult)*3)//4
K = 20
hps = 20
#As we are running in steps of 0.1, need hps*10 positions in the array
hp_errors = np.zeros(hps)
for hp in range(0,hps,1):
    error = 0
    #Run K experiments for each hp value
    for i in range(K):
        #Shuffle indices for random selection.
        np.random.shuffle(tr_ind_mult)
        cur_tr_ind, cur_te_ind = tr_ind_mult[:num_rand], tr_ind_mult[num_rand:]
        cur_tr_data = data[cur_tr_ind]
        cur_te_data = data[cur_te_ind]
        cur_tr_lab_ohe = ohe_labels[cur_tr_ind]
        cur_te_lab = labels[cur_te_ind]
        #Train using current training data and one hot encoded labels
        w = l2_rls_train(cur_tr_data, cur_tr_lab_ohe, hp/10)
        #Set up array to hold predictions
        preds = np.zeros(len(cur_te_data))
        for k in range(len(cur_te_data)):
            #Predict and use argmax + 1 to select prediction as label 1 is held at position 0.
            pred = l2_rls_predict(w, cur_te_data[k,:])
            prediction = np.argmax(pred) + 1
            preds[k] = prediction
        #Add error for this experiment to cumulative error
        correct = np.sum(preds == cur_te_lab)
        error += 1- (correct/cur_te_lab.size)
    #Calculate cumulative error and store in hp_errors
    error = error/K
    hp_errors[hp] = error
print("Overall Errors for different hp values: {}".format(hp_errors))
#As we run in steps of 0.1, value is index/10
hp_val = np.argmin(hp_errors)/10
print("Selected hyperparameter value: {}".format(hp_val))
```

```python
tr_data_mult = data[tr_ind_mult]
#Get ohe labels for training
tr_labels_ohe = ohe_labels[tr_ind_mult]
te_data_mult = data[te_ind_mult]
te_lab_mult = labels[te_ind_mult]
#Train using ohe labels and initialise array to store predictions
weights = l2_rls_train(tr_data_mult, tr_labels_ohe, hp_val)
hp_predictions = np.zeros(len(te_data_mult))
incorrects = []
for i in range(len(te_data_mult)):
    #Predict over every item and use argmax to calculate prediction.
    predict = l2_rls_predict(weights, te_data_mult[i,:])
    hp_prediction = np.argmax(predict) + 1
    hp_predictions[i] = hp_prediction
    #Store incorrectly predicted indices for later use
    if hp_prediction != te_lab_mult[i]:
        incorrects.append(i)

#Build confusion matrix for every 40 images
confusion = np.zeros((40,40))
for i in labels:
    for j in labels:
        confusion[i-1][j-1] = np.count_nonzero((te_lab_mult == i) & (hp_predictions == j))
        

#Output total error
correct = np.sum(hp_predictions == te_lab_mult)
error = 1 - correct/te_lab_mult.size
print("Overall Error: {}".format(error))
```

 Pick some subjects that are considered the easist to recognise by your classifier, and print the test images from these subjects below.

```python
errors = np.zeros(40)
k = 3
#Use confusion matrix to calculate how many are wrongly classified for each classifier.
for j,row in enumerate(confusion):
    total = np.sum(row)
    wrong = total - row[j]
    errors[j] = wrong / total

#Argpartition to get easiest to predict and show the test images
easiest = np.argpartition(errors, k)
print("Some of the easiest to recognise: \n")
for idx in easiest[:k]:
    print(f"Person {idx}")
    for i in range(5):
        show_single_face(te_data_mult[idx*5 + i,:])
```

 Pick some subjects that are considered the most difficult to recognise by your classifier, and print the misclassified test images from these subjects below.

```python
#Use argpartition to calculate hardest to predict.
hardest = np.argpartition(errors, -k)
print("Misclassified test images from hardest subjects: \n")
hard_ind = hardest[-k:]
#Use incorrect array from eariler to print out all images wrongly classified who have a label in hard_ind
for wrong in incorrects:
    if te_lab_mult[wrong] in [x + 1 for x in hard_ind]:
        show_single_face(te_data_mult[wrong,:])

```

## 4. Gradient descent for training linear least squares model

Now, implement the function `lls_gd_train` below. It should train a single-output linear model by minimising the sum of squares loss using the gradient descent approach. It should take the iteration number $N$ and the learning rate $\eta$ as the input. To keep things simple, you can fix the initial guess of the model weights $\mathbf{w}$ as zeros. It should record and return the weights and costs of linear least squares model calculated in all the iterations.

```python
def lls_gd_train(data, labels, learning_rate, N # Add any other arguments here if needed
          ):
    """
    Train linear least squares using gradient descent.
    Over N iterations, calculate cost and gradient at each iteration, 
    then update weights based on learning rate * gradient.

    data: Training data matrix of type np.array 
    labels: Vector of labels in a np.array

    Returns: np vector holding cost at each iteration and a np.array holding weights at each iteration
    """
    
    # This is just to be consistent with the lecture notes.
    X, y = data, labels
    
    # Expand X with a column of ones.
    Xtilde = np.c_[np.ones(X.shape[0]), X]
    
    # Weight initialization
    w = np.zeros(Xtilde.shape[1])
    w_all = np.zeros((N, w.size))
    cost_all = np.zeros(N)
    # GD update of weights
    for i in range(N):
        cost = 0
        # cost and gradient descent of the linear least squares model
        gd = np.transpose(Xtilde)@Xtilde@w - np.transpose(Xtilde)@y
        preds = Xtilde@w
        #Caclulate cost using sum-of-squares loss
        for j,pred in enumerate(preds):
            cost += (pred - y[j])**2
        
        cost = cost / 2
        
        # Weight update
        w = w - learning_rate * (gd)
    
        # save w and cost of each iteration in w_all and cost_all 
        w_all[i] = w
        cost_all[i] = cost
    
    # Return model parameters.
    return cost_all, w_all
```

Apply this function to classify images from subject 1 and subject 30. You only need to conduct the experiment for one training-test trial, where 3 images from each subject are used for training and the remaining 7 images for testing.

Set your learning rate small, e.g., $\eta=10^{-3}$, and use a reasonable iteration number, e.g., $N=200$, to train your model. Plot in three figures (1) the change of your sum-of-squares error loss that you minimise, (2) the change of the classification accuracy for training samples, and (3) the change of the classification accuracy for test samples over iterations. Write your code below and make sure it compiles. 

```python
#Set learning rate and num_iterations
lr = 0.001
N = 200
#Setup array of 1 ... n for graph plotting
ns = np.arange(N)
#Setup data and labels for examples 1 and 30 as we are doing binary classification with these subjects
tr_ind_gd, te_ind_gd = partition_data(labels, num_per_class=3)
tr_ind_gd = tr_ind[(labels[tr_ind_gd]==1)| (labels[tr_ind_gd]==30)]
te_ind_gd = te_ind[(labels[te_ind_gd]==1)| (labels[te_ind_gd]==30)]
tr_labels_gd = labels[tr_ind_gd]
te_labels_gd = labels[te_ind_gd]
tr_data_gd = data[tr_ind_gd, :]
te_data_gd = data[te_ind_gd, :]
tr_labels_gd = np.where(tr_labels_gd==1,-1, tr_labels_gd)
tr_labels_gd = np.where(tr_labels_gd==30, 1, tr_labels_gd)
te_labels_gd = np.where(te_labels_gd==1,-1, te_labels_gd)
te_labels_gd = np.where(te_labels_gd==30, 1, te_labels_gd)
#Calculate costs and weights at each iteration from gradient descent.
costs_all, weights_all = lls_gd_train(tr_data_gd, tr_labels_gd, lr, N)
#Initialise arrays to hold training and testing accuracy at each iteration
class_te_acc = np.zeros(N)
class_tr_acc = np.zeros(N)
N2 = len(tr_data_gd)
N3 = len(te_data_gd)
for i in range(N):
    #Get current weights
    cur_weights = weights_all[i]
    te_acc = 0
    tr_acc = 0
    #Predict and calculate accuracy for training data
    for j in range(N2):
        pred = l2_rls_predict(cur_weights, tr_data_gd[j,:])
        if pred < 0:
            if tr_labels_gd[j] == -1:
                tr_acc += 1
        else:
            if tr_labels_gd[j] == 1:
                tr_acc += 1
    #Predict and calculate accuracy for testing data
    for j in range(N3):
        pred = l2_rls_predict(cur_weights, te_data_gd[j,:])
        if pred < 0:
            if te_labels_gd[j] == -1:
                te_acc += 1
        else:
            if te_labels_gd[j] == 1:
                te_acc += 1
    te_acc = te_acc / N3
    tr_acc = tr_acc / N2
    #Set training and testing accuracy for iteration i to the current value.
    class_te_acc[i] = te_acc
    class_tr_acc[i] = tr_acc
```

```python
plt.plot(ns, costs_all, label="sum-of-squares error")
plt.xlabel("N")
plt.ylabel("Sum-of-Squares Error")
plt.title("Sum-of-Squares Cost at each Iteration")
plt.figure()

plt.plot(ns, class_tr_acc)
plt.xlabel("N")
plt.ylabel("Training classification accuracy")
plt.title("Training classification accuracy at each Iteration")
plt.figure()

plt.plot(ns, class_te_acc)
plt.xlabel("N")
plt.ylabel("Testing classification accuracy")
plt.title("Testing classifcation accuracy at each Iteration")
plt.figure()

plt.show()
```
